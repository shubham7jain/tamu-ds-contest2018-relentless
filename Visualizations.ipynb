Graph of companies vs tripCount for all days

```python
df['Trip Start Month'] = df['Trip Start'].dt.month
graph2 = pd.DataFrame({'tripCount' : df.groupby(['Company', 'Trip Start Month']).size()}).reset_index()
```

```python
graph2['Company'].unique()
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

import pandas as pd

data = []
for i in range(-1, 59):
    company = go.Scatter(
        x=graph2[graph2['Company'] == i]['Trip Start Month'],
        y=graph2[graph2['Company'] == i]['tripCount'],
        name = "Company " + str(i),
        opacity = 0.8)
    data.append(company)

py.iplot(data, filename='pandas-time-series')
```
